```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
working=[5,6,3,7,8]
sleeping=[4,5,7,2,8]
playing=[7,8,3,1,9]
plt.plot([],[],label='Working',color='r',linewidth=5)
plt.plot([],[],label='Sleeping',color='g',linewidth=5)
plt.plot([],[],label='Playing',color='b',linewidth=5)
plt.stackplot(x,working,sleeping,playing,colors=['r','g','b'])
plt.legend()
```
