```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
m=Basemap(projection='mill',resolution='c')
NYCLat,NYCLon=40.7127,-74.0059
m.drawcoastlines()
m.drawcountries()
m.drawstates()
xs,ys=[],[]
xpt,ypt=m(NYCLon,NYCLat)
xs.append(xpt)
ys.append(ypt)
m.plot(xpt,ypt,'c*',markersize=10)
LALat,LALon=34.05,-118.25
xpt,ypt=m(LALon,LALat)
m.plot(xpt,ypt,'c*',markersize=10)
xs.append(xpt)
ys.append(ypt)
m.plot(xs,ys,label='nice')
plt.show()
```
