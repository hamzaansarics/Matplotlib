```python
import matplotlib.pyplot as plt
import numpy as np
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
years=[20,40,10,12,22]
prices=[40,19,10,20,16]
pr=np.array(prices)
yr=np.array(years)
# plt.plot(yr,pr)
m=max(prices)
y=max(years)
for label in ax.yaxis.get_ticklabels():
    label.set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
boxstyle=dict(boxstyle='round4',fc='w',ec='k')
ax.annotate(str(pr[0]),(pr[0],yr[0]),xytext=(m+2,y),bbox=boxstyle)
ax.annotate(str(35),(pr[0],yr[0]),xytext=(m+2,35),bbox=boxstyle)
ax.annotate(str(30),(pr[0],yr[0]),xytext=(m+2,30),bbox=boxstyle)
ax.annotate(str(25),(pr[0],yr[0]),xytext=(m+2,25),bbox=boxstyle)
ax.annotate(str(20),(pr[0],yr[0]),xytext=(m+2,20),bbox=boxstyle)
ax.annotate(str(15),(pr[0],yr[0]),xytext=(m+2,15),bbox=boxstyle)
ax.annotate(str(10),(pr[0],yr[0]),xytext=(m+2,10),bbox=boxstyle)
plt.subplots_adjust(hspace=600)
plt.show()
```
