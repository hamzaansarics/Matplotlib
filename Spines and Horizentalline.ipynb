```python
import matplotlib.pyplot as plt
import numpy as np
years=[2000,2004,2008,2012,2016,2020]
prices=[1000,6000,3000,5000,2000,7000]
fig=plt.figure()
ax=plt.subplot2grid((1,1),(0,0))
pr=np.array(prices)
yr=np.array(years)
ax.fill_between(yr,pr,pr[2],facecolor='g',alpha=0.5)
ax.xaxis.label.set_color('g')
ax.yaxis.label.set_color('r')
ax.set_xticks([2000,2006,2012,2020])
ax.set_yticks([1000,3000,5000,7000])
#Spines
ax.spines['left'].set_color('g')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
#Horizental Line
ax.axhline(pr[2],color='c',linewidth=5)
#--------------------------------------
for label in ax.xaxis.get_ticklabels():
    label.set_color('g')
for label in ax.xaxis.get_ticklabels():
    label.set_rotation(45)
plt.xlabel('Years')
plt.ylabel('Prices')
plt.title('Simple Price up & down Graph')
# ax.legend()
plt.show()
```
